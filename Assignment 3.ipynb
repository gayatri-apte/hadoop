```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv('./CAR.csv')
cars
```

```python
sns.countplot(x=cars['fuel'],hue=cars['transmission'])
```

### Countplot
In the second hand market, there is a high availability of Diesel cars in the given data. Petrol is a close second. CNG & LPG Cars are only a fraction of Petrol and diesel, while electric cars are even lesser. Manual seems to be the top transmission choice. Automatics are popular only in Petrol and Diesel

```python
sns.boxplot(data=cars,x=cars['selling_price'],y=cars['seller_type'])
```

### Boxplot of selling price vs Sellers
We can see that most car prices are contained between 0 and 2,00,000 Rupees. The average cost of a car sold by a trustmark dealer isa higher than one sold by a dealer, which is costlier than indivual sellers. Dealer sold cars have a lot of outliers on the higher side

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=cars,x='year',y='km_driven',hue='fuel')
plt.show()
```

### Scatterplot
This scatterplot explores the realtionship between kilometers, years and fuel. The kilometers rise from the early 2000s, and dip after early 2010s. Diesel cars are mostly highly driven, except for early 2000s and late 2010s/early 2020s

```python
plt.figure(figsize=(10,8))
sns.histplot(data=cars,x='year',hue='fuel')
plt.show()
```

### Histogram
Out of all the cars for sale, we see there is a high amount of cars made/registered in the 2010s. Diesel cars were prefered in that era, with petrol being the top availability in other cases

```python
sns.kdeplot(data=cars,x="selling_price")
```

### Kernel Density of Selling Price
We see the probability distrbution function of the selliong price. If we pick randomly, we have a higher chance of getting a car which is currently sold between INR 0 and INR 2,00,000

```python
plt.figure(figsize=(10,8))
sns.lineplot(data=cars,x='year',y='selling_price',hue='owner')
plt.show()
```

### Lineplot
Here we have compared selling price to model year of cars, ordered by owner. Except for first owner and test drive cars, all increase till late 2010s after which we see a dip. First owner car prices are increasing for model years upto 2020. We see a weird U/V shape line for Test drive cars

```python
plt.figure(figsize=(10,5))
sns.violinplot(data=cars,x='km_driven',y='owner')
plt.show()
```

### Violinplot
We can see that on an average, fourth and above owner cars are driven more. However, first owner cars have a larger probability distribution.The interquartile distance is equal in second owner cars, but quartiles differ for others. First owner cars have the highest amount of outliers in terms of km driven

```python
sns.pairplot(cars,hue='seller_type')
```

### Pairplot
We can infer multiple things from the pairplot
1. Individual owner statistics are widely distributed. Dealers and trustmark dealer only sell cars that fall in some rules
2. Dealers charge higher than individual owners. This may be due to commision or work they do on cars
